## ***Cleaning Our Data***

```python
# Additional Dependeics
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import pandas as pd
import numpy as np
```

```python
# This is the file import and you can upload it when you file run it so you don't have to reload each time.
data = 'tester.txt'
df = pd.read_csv(data)
```

```python
# Preview of the DataFrame
df
```

```python
df = df.loc[df['Survival months'] != 'Unknown']
```

```python
df['Survival months'] = pd.to_numeric(df['Survival months'])
df['Survival months'].dtype
```

```python
df.columns
```

```python
# Rename the columns in new data frame
df_1 = df.rename(columns={'Sex': 'Gender', 'Age recode with single ages and 85+':'Age',
       'Median household income inflation adj to 2018':'Median Household Income',
       'Primary Site - labeled':'Cancer Site', 'Histology recode - broad groupings':'Cancer Type',
       'Derived AJCC Stage Group, 7th ed (2010-2015)':'Cancer Stage', 'Survival months':'Survival Months',
       'Race and origin recode (NHW, NHB, NHAIAN, NHAPI, Hispanic)':'Race'})
df_1
```

```python
df_1['Age'] = df_1['Age'].str.replace(' years','')
df_1['Age'] = df_1['Age'].str.replace('+','')
df_1
```

```python
# # Rename the columns in new data frame
# df_1 = df.rename(columns={'Survival months':'Survival Months','Age recode with <1 year olds': 'Age Group',
#        'Derived AJCC Stage Group, 7th ed (2010-2015)':'Cancer Stage', 'Sex': 'Gender',
#        'CS Schema - AJCC 6th Edition': 'Cancer Site',
#        'Race and origin recode (NHW, NHB, NHAIAN, NHAPI, Hispanic)':'Race',
#        'Median household income inflation adj to 2018':'Median Household Income'})
# df_1
```

```python
df_1 = df_1.loc[df_1['Age'] != 'Unknown']
```

```python
df_1['Age'] = pd.to_numeric(df_1['Age'])
df_1['Age'].dtype
```

```python
# Drop all rows with missing information
df_clean_data = df_1.dropna(how='any')
```

```python
#Iterate through rows to identify what columns have null values
for column in df_clean_data.columns:
    print(f"Column {column} has {df_clean_data[column].isnull().sum()} null values")
```

```python
df_clean_data.head()
```

### ***Cleaned Data Frame***

```python
y = df_clean_data['Survival Months']
y
```

```python
testing_data = df_clean_data.drop('Survival Months', axis=1)
testing_data
```

```python
X = pd.get_dummies(testing_data)
X
```

### ***Start of Machine Learning***

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
model = LinearRegression()
model
```

```python
model.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {model.score(X_train, y_train)}")
print(f"Testing Data Score: {model.score(X_test, y_test)}")
```

```python
model.coef_
```

```python
model = RandomForestRegressor()
model
```

```python
model.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {model.score(X_train, y_train)}")
print(f"Testing Data Score: {model.score(X_test, y_test)}")
```

```python
predicated = model.predict(X_test)
mean_absolute_error(y_test,predicated)
```
